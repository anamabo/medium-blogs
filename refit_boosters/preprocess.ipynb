```python
import os
import pandas as pd
```

```python
full_dataset = pd.read_csv('./data/hearth_2022/heart_2022_no_nans.csv')
```

```python
full_dataset.head()
```

```python
int_cols = ['PhysicalHealthDays', 'MentalHealthDays', 'SleepHours']
float_cols = ['HeightInMeters', 'WeightInKilograms', 'BMI']
bool_cols = ['PhysicalActivities', 'HadAngina', 'HadStroke', 'HadAsthma', 'HadSkinCancer', 'HadCOPD', 'HadDepressiveDisorder', 'HadKidneyDisease', 'HadArthritis', 'DeafOrHardOfHearing', 'BlindOrVisionDifficulty', 'DifficultyConcentrating', 'DifficultyDressingBathing', 'DifficultyDressingBathing', 'DifficultyErrands', 'ChestScan', 'AlcoholDrinkers', 'HIVTesting', 'FluVaxLast12', 'PneumoVaxEver', 'HighRiskLastYear']
target = ['HadHeartAttack']  # bool
cat_cols = list(set(full_dataset.columns) - set(int_cols) - set(float_cols) - set(bool_cols) - set(target))
```

```python
full_dataset[int_cols] = full_dataset[int_cols].astype(int)
full_dataset[float_cols] = full_dataset[float_cols].astype(float)
full_dataset[cat_cols] = full_dataset[cat_cols].astype('category')
```

```python
for col in bool_cols + target:
    full_dataset[col] = (full_dataset[col] == 'Yes')
```

```python
os.makedirs('./data/preprocess_data/', exist_ok=True)
```

```python
full_dataset.to_parquet('./data/preprocess_data/heart_2022.parquet')
```
